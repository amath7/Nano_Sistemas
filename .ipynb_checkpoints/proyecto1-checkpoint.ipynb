vimos en clase que la estructura de bandas para la **cadena infinita** (el cero de la energía como $\epsilon_0 = 0$) es de la forma:
\begin{eqnarray}
\epsilon (k) = \epsilon_0 + 2\cos(k)
\end{eqnarray}
Para la densidad de energía se tiene:
\begin{eqnarray}
D(E) = \dfrac{\Theta \left( 1 - \vert \dfrac{E}{2} \vert \right)}{2\pi \sqrt{1 - \left(\dfrac{E}{2} \right)^2}}
\end{eqnarray}


```julia
Θ(x) = (x < 0) ? 0. : 1.
DOSchain(E) = Θ(1 - abs(E/2)) / (2pi * sqrt(1 - (E/2)^2))
```

```julia
using Plots
ϵ0 = 0
k = linspace(-pi, pi, 100)
E = linspace(-2, 2, 100)
p1 = plot(k, (ϵ0 .+ 2.*cos.(k)), xlabel="k", ylabel="E(k)")
p2 = plot(E, DOSchain.(E), xlabel="E", ylabel="D(E)", xlims=(-2, 2), ylims=(0, 0.8))
plot(p1, p2, size=(1000, 500))
```

la estructura de bandas para la **malla** (el cero de la energía como $\epsilon_0 = 0$) es de la forma:
\begin{eqnarray}
\epsilon (q, k) = 2\left( \cos(q) + \cos(k) \right)
\end{eqnarray}
Para la densidad de energía se tiene:
\begin{eqnarray}
D(E) = \dfrac{1}{2pi^2} \kappa \left( 1 - \left( E/4 \right) \right) \Theta \left( 1 - |E/4| \right)
\end{eqnarray}
con
\begin{eqnarray}
\kappa \left( x \right) = \int_{0}^{\pi /2} d\phi \left(1 - x\left(\sin(\phi)\right)^2 \right)^{-1/2}
\end{eqnarray}

Por otro lado la densidad de los estados para un número finito de sitios, es de la forma:
\begin{eqnarray}
D_M(E) = \dfrac{1}{2\pi M} \sum_{j=1}^{M} \dfrac{\Theta\left(1 - \lvert E/2 - \cos\left(\dfrac{\pi j}{M+1}\right)\rvert\right)}{\sqrt{1 - \left(E/2 - \cos\left(\dfrac{\pi j}{M+1}\right)\right)^2 }}
\end{eqnarray}

```julia
function riemmanSum(f,a,b,N=1000)
    δ = 1/N
    sum = 0
    for i in a:δ:b
        sum += f(i)*δ
    end
    return sum
end
```

```julia
f(ϕ, x) = (1-x*(sin(ϕ))^2)^(-1/2)
κ(x) = riemmanSum(ϕ -> f(ϕ, x), 0, pi/2)
Eribbon(q, k) = 2 * (cos(k) + cos(q))
DOSribbon(E) = (sqrt(2)*pi)^(-2) * κ(1-(E/4)^2) * Θ(1 - abs(E/4))
```

```julia
function DOSribbonM(E::Float64, M::Int64)
    sum = 0
    for j in 1:M
        sum += real(Θ(1 - abs(E/2 - cos(pi*j/(M+1)))) / (sqrt(complex(1 - (E/2 - cos(pi*j/(M+1)))^2))) )
    end
    return (1/2pi*M) * sum
end
```

```julia
E = linspace(-4, 4, 1000)
plot(E, DOSribbon.(E), xlabel="E", ylabel="D(E)", label="N >> 1, M >> 1")
```

```julia
plot(E, DOSribbonM.(E, 3), xlabel="E", ylabel="D(E)", label="N >> 1, M = 3")
```

```julia
E = linspace(-4, 4, 1000)
M = 3
plot(E, DOSribbonM.(E, M), xlabel="E", ylabel="D(E)", label="N >> 1, M = 3")
```
